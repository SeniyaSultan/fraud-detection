```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load processed fraud data
fraud_clean_path = r"C:\Users\jkk\OneDrive\Desktop\fraud-detection\data\processed\fraud_clean.csv"
fraud_df = pd.read_csv(fraud_clean_path)

# Separate features and target
X = fraud_df.drop('class', axis=1)
y = fraud_df['class']

# Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Save to processed folder
X_test.to_csv(r"C:\Users\jkk\OneDrive\Desktop\fraud-detection\data\processed\X_test.csv", index=False)
y_test.to_csv(r"C:\Users\jkk\OneDrive\Desktop\fraud-detection\data\processed\y_test.csv", index=False)
X_train.to_csv(r"C:\Users\jkk\OneDrive\Desktop\fraud-detection\data\processed\X_train.csv", index=False)
y_train.to_csv(r"C:\Users\jkk\OneDrive\Desktop\fraud-detection\data\processed\y_train.csv", index=False)
```
